```python
# 날짜 변환 함수 정의
def convert_to_date_only(date_str):
    try:
        # 포맷 없이 변환 시도
        return pd.to_datetime(date_str).date()
    except Exception:
        try:
            # 두 번째 형식 처리: "24.12.06."
            return pd.to_datetime(date_str, format='%y.%m.%d.').date()
        except Exception as e:
            print(f"Error converting {date_str}: {e}")
            return pd.NaT  # 변환 실패 시 NaT 반환
```

```python
import pandas as pd
import os
from glob import glob

# 1. 모든 하위 폴더에서 CSV 파일 경로 수집
root_dir = "/content/drive/MyDrive/DIC/블로그"
csv_files = glob(os.path.join(root_dir, "**/*.csv"), recursive=True)


# CSV 파일들을 하나의 리스트로 읽기
df_list = []
for file in csv_files:
    data = pd.read_csv(file)
    print(file)
    data["date"] = data["date"].astype(str)
    # 데이터프레임에 변환 적용
    data['date'] = data['date'].apply(convert_to_date_only)
    data = data.dropna(subset=['date'])
    # 기준 날짜로 필터링
    data = data[data['date'].dt.date >= pd.to_datetime('2020-01-01').date()].reset_index(drop=True)
    df_list.append(data)

# 모든 필터링된 데이터프레임을 하나로 합치기
df = pd.concat(df_list, axis=0).reset_index(drop=True)

# 결과 확인
print(df.head())
```

## \n 제거

```python
def preprocess_text(text):
    if isinstance(text, str):
        text = re.sub(r'<[^>]+>', '', text)
        text = re.sub(r'\n', ' ', text)
        text = re.sub(r'\s+', ' ', text)

    return text
```

```python
df['title'] = df['title'].apply(preprocess_text)
df['content'] = df['content'].apply(preprocess_text)
df['date'] = df['date'].apply(preprocess_text)
```
